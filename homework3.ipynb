```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
print(data)
plt.plot(data[:,0],data[:,1: ])
plt.legend(['Hare','Lynx','Carrot'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population')
plt.show()
```

№2

```python
plt.plot(data[:,0],data[:,1: ])
biger1 = np.where(data[:, 1] > 50000)
biger2 = np.where(data[:, 2] > 50000)
biger3 = np.where(data[:, 3] > 50000)
plt.vlines(data[biger1, 0], 0, data[biger1, 1], linestyles="dashed")
plt.vlines(data[biger2, 0], 0, data[biger2, 2], linestyles="dashed")
plt.vlines(data[biger3, 0], 0, data[biger3, 3], linestyles="dashed")
plt.plot(data[biger1, 0], data[biger1, 1] , 'gs',c = 'g')
plt.plot(data[biger2, 0], data[biger2, 2] ,'gs', c = 'g')
plt.plot(data[biger3, 0], data[biger3, 3] ,'gs', c = 'g')

plt.legend(['Hare','Lynx'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('more 50000')
plt.grid()
plt.show()
```

№3

```python
plt.plot(data[:,0],data[:,1: ])
plt.vlines(data[np.argmax(data[:, 1]), 0], 0, data[np.argmax(data[:, 1]), 1], linestyles="dashed")
plt.vlines(data[np.argmax(data[:, 2]), 0], 0, data[np.argmax(data[:, 2]), 2], linestyles="dashed")
plt.vlines(data[np.argmax(data[:, 3]), 0], 0, data[np.argmax(data[:, 3]), 3], linestyles="dashed")
plt.plot(data[np.argmax(data[:, 1]), 0],data[np.argmax(data[:, 1]), 1],'gs',c='r')
plt.plot(data[np.argmax(data[:, 2]), 0],data[np.argmax(data[:, 2]), 2],'gs',c='r')
plt.plot(data[np.argmax(data[:, 3]), 0],data[np.argmax(data[:, 3]), 3],'gs',c='r')
plt.legend(['Hare','Lynx','Carrot'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Max Population')
plt.grid()
plt.show()
```

№4

```python
plt.title('histogram')
all = data[:,1]+data[:,2]
hare = data[:,1]/all*100
lynx = data[:,2]/all*100
plt.bar(data[:, 0], label='Hare', width = 1, height = hare, color = "brown")
plt.bar(data[:, 0], label='Lynx', width = 1, height = lynx, bottom = hare, color = 'blue')
plt.legend(['Hare','Lynx'])
plt.grid()
plt.show()
```

№5

```python
b, a = plt.subplots(1, 3,sharey=True)
b.suptitle('mans and std')
a[0].barh(data[:, 0], data[:, 1], label='Hare', color = "blue")
a[1].barh(data[:, 0], data[:, 2], label='Lynx', color = "yellow")
a[2].barh(data[:, 0], data[:, 3], label='Carrot', color ='g')
for i in range(0,3):
  a[i].vlines(np.mean(data[:, i+1]), 1899, 1921, colors = 'r')
  a[i].vlines(np.std(data[:, i+1]), 1899, 1921, colors = 'black')
  a[i].legend()
  a[i].grid()
plt.show()
```

№6

```python
s = np.where(data[:, 1] - data[:, 2] > 0, 'H', 'L')
pc = np.array([len(np.argwhere(s == 'H')), len(np.argwhere(s == 'L'))])
print(s)
plt.pie(pc, labels=(['Hare', 'Lynx']), colors = ['red','blue'])
plt.show()
```

```python
new = np.insert(np.diff(data[:, 1]), 0, 0)
plt.plot(data[:, 0], new, label='Hare', c = 'g')
plt.plot(data[:,0],data[:,2])
plt.legend(['Hare','Lynx'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Change')
plt.grid()
plt.show()
```

№8

```python
b, a = plt.subplots(1, 2, sharey=True)
b.suptitle('Hares vs Carrots')
a[0].plot(data[:, 0], data[:, 1], label='Hare', color = "purple")
a[1].plot(data[:, 0], data[:, 3], label='Carrot', color="orange")
for i in range(0,2):
  a[i].legend()
  a[i].grid()
plt.show()
b, a = plt.subplots(1, 2, sharey=True)
b.suptitle('Hares vs Lynxes')
a[0].plot(data[:, 0], data[:, 1], label='Hare', color = "purple")
a[1].plot(data[:, 0], data[:, 2], label='Lynxes', color = "orange")
for i in range(0,2):
  a[i].legend()
  a[i].grid()
plt.show()
```
